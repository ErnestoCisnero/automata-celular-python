```python
import numpy
import matplotlib.pyplot as plt 
```

```python
number_rule = 30
output = [int(x) for x in numpy.binary_repr(number_rule, width=8)] #Representacion binaria del numero que representa la regla
input_pattern = numpy.zeros([8,3]) #Matriz de 8x3, cada fila representa un posible estado de la celula
for i in range(8):
    input_pattern[i,:] = [int(x) for x in numpy.binary_repr(7-i, width=3)] #Se asigna a cada fila un valor binario (0-7) que representa los posibles estados a partir de 3 celulas                                                                               #para asignar el valor futuro de la siguiente generacion de celula.
columns = 24
rows = int(columns/2)
canvas = numpy.zeros([rows,columns+2])
canvas[0,int(columns/2)] = 1

for i in range(0,rows-1):
    for j in range(0,columns):
        for k in range(8):
            if numpy.array_equal(input_pattern[k,:],canvas[i,j:j+3]):
                canvas[i+1,j+1] = output[k]


plt.imshow(canvas[:,1:columns+1], cmap='Greys', interpolation='none')
plt.title("Autómata Celular de regla {}".format(number_rule))
plt.show()
```
